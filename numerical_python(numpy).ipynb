Concatenating Arrays vertically and horizontally

```python
arr1=np.arange(12,24).reshape(3,4)
print(arr1)
```

```python
arr2=np.arange(0,12).reshape(3,4)
print(arr2)
```

```python
vertical_concat=np.vstack((arr1,arr2))
print(vertical_concat)
```

```python
vertical_concat=np.vstack((arr1+arr2))
print(vertical_concat)
```

```python
split_arr=np.split(arr1,2,axis=1)
print(split_arr)
```
